```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set display option to avoid scientific notation in pandas, show up to 5 decimal points
pd.set_option('display.float_format', lambda x: '%.5f' % x)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 120)
pd.set_option('display.float_format', lambda x: f'{x:.4f}')
```

```python
# Load the dataset 
flood_data = pd.read_excel("flood_full_data.xlsx")
flood_data.head()

```

```python
flood_data.columns
```

```python
flood_data['Start Date'] = pd.to_datetime(flood_data[['Start Year', 'Start Month', 'Start Day']].rename(columns={'Start Year': 'year', 'Start Month': 'month', 'Start Day': 'day'}), errors='coerce')
flood_data['End Date'] = pd.to_datetime(flood_data[['End Year', 'End Month', 'End Day']].rename(columns={'End Year': 'year', 'End Month': 'month', 'End Day': 'day'}), errors='coerce')
flood_data['Duration'] = (flood_data['End Date'] - flood_data['Start Date']).dt.days.astype('Int64')
flood_data.drop(columns=['Start Year', 'Start Month', 'Start Day', 'End Year', 'End Month', 'End Day'], inplace=True)
#order flood by date soonest first
flood_data.sort_values(by='Start Date', inplace=True, ascending=False)
#reset index
flood_data.reset_index(drop=True, inplace=True)
#drop rows with NaT in start and end date
flood_data.dropna(subset=['Start Date', 'End Date'], inplace=True)
```

```python

#create ID columns
flood_data['Flood_ID'] = flood_data.index + 1
#make flood id fisrt collumns 
cols = flood_data.columns.tolist()
flood_data = flood_data[cols[-1:] + cols[:-1]]

#first 150 rows
flood_extra_data = flood_data.iloc[:150].copy()

flood_extra_data.head()
```

```python
flood_extra_data.shape
```

```python
# Convert columns to appropriate data types
flood_extra_data['Flood_ID'] = flood_extra_data['Flood_ID'].astype('int64')
flood_extra_data['ISO'] = flood_extra_data['ISO'].astype('string')
flood_extra_data['Country'] = flood_extra_data['Country'].astype('string')
flood_extra_data['Subregion'] = flood_extra_data['Subregion'].astype('string')
flood_extra_data['Region'] = flood_extra_data['Region'].astype('string')
flood_extra_data['River Basin'] = flood_extra_data['River Basin'].astype('string')
flood_extra_data['Duration'] = flood_extra_data['Duration'].astype('Int64') 

#drop all collumns with  50 non-null values
drop_cols = flood_extra_data.columns[flood_extra_data.notnull().sum() < 50]
flood_extra_data = flood_extra_data.drop(columns=drop_cols)
#drop entry date and last update date
flood_extra_data = flood_extra_data.drop(columns=['Entry Date', 'Last Update'])
```

```python
relevant_columns=['Flood_ID','Disaster Type','Disaster Subtype','ISO', 'Country', 'Subregion', 'Region', 'Location', 'Origin','Associated Types','River Basin', 
                  'Start Date', 'End Date','Duration']

flood_pd=flood_data[relevant_columns].copy()

#reset index
flood_pd.reset_index(drop=True, inplace=True)
#create ID columns

flood_pd.head()


```

```python
#flood for common crawl last 150 events
rel_columns=['Flood_ID','ISO', 'Country', 'Subregion', 'Region', 'Location', 'River Basin', 'Start Date', 'End Date','Duration']
#use relevant columns
flood_crawl= flood_pd[rel_columns].iloc[:150].copy()

flood_crawl.columns
```

```python
flood_with_lang=pd.read_csv('flood_crawl_with_languages.csv')

#'Language_ISO_639_3', 'Local_Languages' add this columns to flood_crawl
flood_crawl['Language_ISO_639_3'] = flood_with_lang['Language_ISO_639_3']
flood_crawl['Local_Languages'] = flood_with_lang['Local_Languages']

#order of the collumns 
#save to csv
flood_crawl.to_csv('../Data/Flood_Data/flood_crawl.csv', index=False)

```

```python
# Convert comma-separated strings to lists and strip whitespace
flood_crawl['Location'] = flood_crawl['Location'].apply(lambda x: [loc.strip() for loc in x.split(',')] if pd.notna(x) and isinstance(x, str) else [])
flood_crawl['Language_ISO_639_3'] = flood_crawl['Language_ISO_639_3'].apply(lambda x: [lang.strip() for lang in x.split(',')] if pd.notna(x) and isinstance(x, str) else [])
flood_crawl['Local_Languages'] = flood_crawl['Local_Languages'].apply(lambda x: [lang.strip() for lang in x.split(',')] if pd.notna(x) and isinstance(x, str) else [])
```

```python
# Convert columns to appropriate data types
flood_crawl['Flood_ID'] = flood_crawl['Flood_ID'].astype('int64')
flood_crawl['ISO'] = flood_crawl['ISO'].astype('string')
flood_crawl['Country'] = flood_crawl['Country'].astype('string')
flood_crawl['Subregion'] = flood_crawl['Subregion'].astype('string')
flood_crawl['Region'] = flood_crawl['Region'].astype('string')
flood_crawl['River Basin'] = flood_crawl['River Basin'].astype('string')
flood_crawl['Duration'] = flood_crawl['Duration'].astype('Int64')  # nullable integer

# Start Date and End Date should already be datetime, but ensure it
flood_crawl['Start Date'] = pd.to_datetime(flood_crawl['Start Date'])
flood_crawl['End Date'] = pd.to_datetime(flood_crawl['End Date'])

# Location, Language_ISO_639_3, and Local_Languages remain as object (lists)
# They're already converted to lists in the previous cell

# Verify the data types
flood_crawl.info()
```
