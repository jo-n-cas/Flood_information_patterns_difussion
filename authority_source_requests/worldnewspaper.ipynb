```python
# First, check what countries are actually in the dataframe
print("Countries in df before grouping:")
print(df['Country'].unique())

# Group by country and sum the Count column
country_totals = df.groupby('Country')['Count'].sum().reset_index()
country_totals.columns = ['Country', 'Total_Newspapers']

# Strip whitespace from country names
country_totals['Country'] = country_totals['Country'].str.strip()

# Sort by total newspapers in descending order
country_totals = country_totals.sort_values('Total_Newspapers', ascending=True)

# Correct the name of the countries 
replacements = {
    "United States of America": "United States",
    "Iran (Islamic Republic of)": "Iran",
    "Bolivia (Plurinational State of)": "Bolivia",
    "Democratic Republic of the Congo": "Congo",
    "Republic of Korea": "South Korea",
    "Taiwan (Province of China)": "Taiwan",
    "China, Hong Kong Special Administrative Region": "Hong Kong",
    "Lao People's Democratic Republic": "Laos",
    "United Republic of Tanzania": "Tanzania",
    "Viet Nam": "Vietnam",
    "Syrian Arab Republic": "Syria",
    "Cabo Verde": "Cape Verde"
}

# Apply the replacements
country_totals['Country'] = country_totals['Country'].replace(replacements)

print("\nCountries in country_totals after replacement:")
print(country_totals['Country'].unique())

# Also apply replacements to List_countries for comparison
List_countries_cleaned = [replacements.get(country, country) for country in List_countries]

# Identify which countries are not in the country_totals DataFrame
missing_countries = [country for country in List_countries_cleaned if country not in country_totals['Country'].values]

print("\nMissing countries:", missing_countries)
```
